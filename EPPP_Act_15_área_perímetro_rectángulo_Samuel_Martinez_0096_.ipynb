📝 Redacción del Problema
Se requiere desarrollar un programa interactivo en Python que permita realizar cálculos geométricos básicos sobre un rectángulo. El usuario debe poder elegir entre calcular el área o el perímetro mediante un menú. El programa debe solicitar la base y la altura, aplicar las fórmulas correspondientes y mostrar los resultados de forma clara hasta que el usuario decida salir.
📐 Fórmulas Utilizadas Para un rectángulo de base $b$ y altura $h$:
Para un rectángulo de base $b$ y altura $h$:
1. Área: $A = b \times h$
2. Perímetro: $P = 2 \times (b + h)$

![texto del vínculo](https://raw.githubusercontent.com/samuelmartinez128/Act-15-area-perimetro-rectamgulo/refs/heads/main/rectangulo.png)


* Donde b representa la base, h la altura de la figura y se utilizan operadores aritméticos estándar para obtener los resultados geométricos.

* Para implementar estos cálculos de manera práctica, se ha desarrollado un programa en Python estructurado de forma modular y enfocado en el rectángulo.

* El programa utiliza un menú interactivo que se mantiene en ejecución mediante un ciclo while, permitiendo al usuario realizar múltiples cálculos de área y perímetro sin necesidad de reiniciar la aplicación.

* La navegación se controla con estructuras condicionales if, elif y else, las cuales dirigen el flujo del programa hacia la función específica según la opción elegida por el usuario.

* Cada operación está encapsulada en su propia función (arearectangulo() y perimetrorectangulo()), lo que cumple con el principio de responsabilidad única. Esto facilita la comprensión del código y su mantenimiento, siendo un ejemplo ideal para estudiantes de nivel principiante.

* Al invocar cada función, el sistema solicita directamente los valores de la base y la altura, aplica la fórmula matemática correspondiente y muestra el resultado con un formato limpio, preciso y decorado con elementos visuales como emojis para mejorar la experiencia del usuario.

* Este enfoque no solo refuerza el pensamiento lógico y la sintaxis básica de Python, sino que también demuestra cómo la programación es una herramienta potente y práctica para resolver problemas matemáticos de forma automatizada y eficiente.

```python
# =====================================================
# 📐 PROGRAMA DE CÁLCULO GEOMÉTRICO: RECTÁNGULO 📐
  # Elaborado por: Samuel Martinez NL = 35
# Calcular Área del Rectángulo
# Fórmula: A = base * altura
# Calcular Perímetro del Rectángulo
# Fórmula: P = 2 * (base + altura)
# =====================================================

def arearectangulo():
    print("\n" + "="*50)
    print("🟩 CÁLCULO DEL ÁREA DEL RECTÁNGULO 🟩")
    print("="*50)
    base = float(input("📏 Ingresa la base del rectángulo: "))
    altura = float(input("📏 Ingresa la altura del rectángulo: "))
    area = base * altura
    print(f"\n✨ RESULTADO:")
    print(f"   Base: {base} | Altura: {altura}")
    print(f"   Área: {area:.2f} unidades cuadradas")
    print("="*50)

def perimetrorectangulo():
    print("\n" + "="*50)
    print("🟩 CÁLCULO DEL PERÍMETRO DEL RECTÁNGULO 🟩")
    print("="*50)
    base = float(input("📏 Ingresa la base del rectángulo: "))
    altura = float(input("📏 Ingresa la altura del rectángulo: "))
    perimetro = 2 * (base + altura)
    print(f"\n✨ RESULTADO:")
    print(f"   Base: {base} | Altura: {altura}")
    print(f"   Perímetro: {perimetro:.2f} unidades")
    print("="*50)

def menu():
    opcion = 0

    while opcion != 3:
        print("\n")
        print("🌟" * 50)
        print("🌟" + " "*46 + "🌟")
        print("🌟" + "   📐 MENÚ DE RECTÁNGULO 📐   ".center(46) + "🌟")
        print("🌟" + " "*46 + "🌟")
        print("🌟" * 50)
        print("\n   1️⃣  Calcular área del rectángulo")
        print("   2️⃣  Calcular perímetro del rectángulo")
        print("   3️⃣  Salir del programa")
        print("\n" + "─"*50)

        opcion = int(input("\n💡 Elige una opción (1-3): "))

        if opcion == 1:
            arearectangulo()
        elif opcion == 2:
            perimetrorectangulo()
        elif opcion == 3:
            print("\n" + "🎉"*25)
            print("\n   👋 ¡Gracias por usar el programa!")
            print("   📚 ¡Sigue practicando matemáticas!")
            print("   💪 ¡Hasta pronto!\n")
            print("🎉"*25 + "\n")
        else:
            print("\n⚠️  Opción no válida. Por favor elige 1, 2 o 3.")
            print("─"*50)

# =====================================================
# PROGRAMA PRINCIPAL
# =====================================================
if __name__ == "__main__":
    print("\n" + "🚀"*25)
    print("\n   🎓 BIENVENIDO AL PROGRAMA GEOMÉTRICO 🎓")
    print("   Cálculo de Área y Perímetro de Rectángulos")
    print("\n" + "🚀"*25)
    menu()
```
